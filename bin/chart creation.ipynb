```python
import pandas as pd
import glob

csv_files = glob.glob(r"/home/beingscorpion/Semester 3/*.csv")

target_name = 's1mple'

total_kills, total_deaths, total_assists, total_adr = 0, 0, 0, 0
total = 0

for file in csv_files:
    df = pd.read_csv(file)
    
    
    print(f"Columns in {file}: {df.columns.tolist()}")
    
    
    df.columns = df.columns.str.strip()
    
    
    if 'Name' not in df.columns:
        print(f"'Name' column not found in {file}, skipping this file.")
        continue
    
    
    df['Name'] = df['Name'].astype(str).str.lower().str.strip()
    filtered_row = df[df["Name"] == target_name]

    if not filtered_row.empty:
        print(f"Found {target_name} in {file} with data:")
        print(filtered_row)

        total_kills += filtered_row["Kills"].iloc[0] 
        total_deaths += filtered_row["Deaths"].iloc[0]  
        total_assists += filtered_row["Assists"].iloc[0]  
        total_adr += filtered_row["ADR"].iloc[0]  
        total += 1

if total > 0:
    avg_kills = total_kills / total
    avg_deaths = total_deaths / total
    avg_assists = total_assists / total
    avg_adr = total_adr / total
    
    print(f"Name: {target_name}")
    print(f"Average Kills: {avg_kills}")
    print(f"Average Deaths: {avg_deaths}")
    print(f"Average Assists: {avg_assists}")
    print(f"Average ADR: {avg_adr}")
    print(f"Total number of records: {total}")

else:
    print(f"No data found for {target_name}")
```
