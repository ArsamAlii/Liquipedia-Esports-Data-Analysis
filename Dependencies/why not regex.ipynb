```python
import pandas as pd
from difflib import get_close_matches
import glob 

def checking(name):
    csv_files = glob.glob(r"/home/beingscorpion/Semester 3/Player Stats/*.csv")
    name = name.lower()  # lowercasing the name input

    for file in csv_files:
        df = pd.read_csv(file)
        df.columns = df.columns.str.strip()

        if 'Name' not in df.columns:
            continue

        df['Name'] = df['Name'].astype(str).str.lower().str.strip()
        filtered_row = df["Name"].tolist()

        close_matches = get_close_matches(name, filtered_row, n=1, cutoff=0.6)
    
        if close_matches:
            return f"Did you mean '{close_matches[0]}'?"
        
    else:
        return "No close match found."

input_word = 'simple'
result = checking(input_word)
print(result)
```

```python
import pandas as pd
import re

def checking(input_word, df_column):

    matches = []
    pattern = re.compile(f".*{re.escape(input_word)}.*", re.IGNORECASE)
    
    for word in df_column:
        if re.search(pattern, word):
            matches.append(word)
    
    if not matches:
        return "No close match found."
    return f"Did you mean: '{matches[0]}'?"

data = {'words': ['s1mple', 'zywoo', 'donK', 'scream']}
df = pd.DataFrame(data)

name = 'simple'  
df_column = df['words'].tolist()  # Convert the column to a list
result = checking(name, df_column)
print(result)
```
